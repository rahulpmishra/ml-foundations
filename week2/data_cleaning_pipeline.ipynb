```python
first_class = df[(df['Survived']==1) & (df['Pclass']==1)]
first_class.shape
```

```python
# isin — use instead of multiple OR conditions

classes_1_and_2 = df[df['Pclass'].isin([1,2])]
classes_1_and_2.shape
```

```python
# ~ means NOT in pandas

not_first_class = df[~df['Pclass'].isin([1])]
print(not_first_class.shape)
```

```python
# WRONG — this will give a SettingWithCopyWarning or silent bug
survivors = df[df['Survived'] == 1]
survivors['NewCol'] = 1   # BAD
```

```python
# RIGHT — always use .copy() when you plan to modify a slice
survivors = df[df['Survived'] == 1].copy()
survivors['NewCol'] = 1   # SAFE
```

### Missing Value Handling

```python
df = pd.read_csv('titanic.csv')
```

```python
# First: understand the missing data

missing_report = pd.DataFrame({
    'column': df.columns,
    'missing_count':df.isnull().sum(),
    'missing_pct': round((df.isnull().sum()/len(df) * 100),2),
    'dtype':df.dtypes
})

missing_report = missing_report[missing_report['missing_count'] > 0]
print(missing_report)
```

##### THREE strategies for missing values:
###### 1. DROP — when missing % is very high (>50%) or column is irrelevant
###### 2. FILL with statistics — when missing % is low and column is numeric
###### 3. FILL with mode/constant — when column is categorical


###### Strategy 1: Drop column 'Cabin' — 77% missing, not worth it


```python
df_clean = df.drop(columns=['Cabin'])
print(f"After dropping Cabin: {df_clean.shape}")
```

###### Strategy 2: Fill 'Age' with median (not mean — why? outliers)


```python
age_median = df_clean['Age'].median()
df_clean['Age'] = df_clean['Age'].fillna(age_median)
print(f"Age nulls remaining: {df_clean['Age'].isnull().sum()}")
```

###### Strategy 3: Fill 'Embarked' with mode (most frequent)


```python
embarked_mode = df_clean['Embarked'].mode()[0]
df_clean['Embarked'] = df_clean['Embarked'].fillna(embarked_mode)
df_clean['Embarked'].isnull().sum()
```

```python
# Verify: no more nulls
print(f"\nTotal nulls remaining: {df_clean.isnull().sum().sum()}")
```

### Method Chaining

```python
df_raw = pd.read_csv('titanic.csv')
```

```python
# procedural style

df = df_raw.copy()

df = df.drop(columns=['Cabin','Ticket','PassengerId'])
df = df.dropna(subset=['Embarked'])
df['Age'] = df['Age'].fillna(df['Age'].median())
df['Title'] = df['Name'].str.extract(r',\s(\w+)\.')
df['Title'] = df['Title'].replace(['Lady', 'Countess', 'Dr', 'Rev', 'Major',
                                    'Col', 'Capt', 'Sir', 'Jonkheer', 'Dona', 'Don'], 'Rare')
df['Sex'] = df['Sex'].map({'male':0,'female':1})
df = df.drop(columns=['Name'])
print(df.shape)
```

```python
# method chaining style

def clean_titanic(filepath: str) -> pd.DataFrame:
    """
    Full Titanic cleaning pipeline as a method chain.
    Raw CSV in -> clean DataFrame out.
    """

    return (
        pd.read_csv(filepath)
        .drop(columns=['Cabin','Ticket','PassengerId'])
        .dropna(subset=['Embarked'])
        .assign(
            Age = lambda df: df['Age'].fillna(df['Age'].median()),
            Sex = lambda df: df['Sex'].map({'male':0, 'female':1}),
            Title = lambda df: df['Name'].str.extract(r',\s(\w+)\.')[0],
            FamilySize=lambda df: df['SibSp'] + df['Parch'] + 1,
            IsAlone=lambda df: (df['SibSp'] + df['Parch'] == 0).astype(int),
        )
        .assign(
            Title=lambda df: df['Title'].replace(
                ['Lady', 'Countess', 'Dr', 'Rev', 'Major',
                 'Col', 'Capt', 'Sir', 'Jonkheer', 'Dona', 'Don'], 'Rare')
        )
        .drop(columns=['Name'])
        .reset_index(drop=True)
    )

df_clean = clean_titanic('titanic.csv')
print(df_clean.shape)
print(df_clean.head())
print(df.isnull().sum())
```

```python
# pipe() — For Functions That Are Too Complex for assign()

def add_fare_bins(df: pd.DataFrame) -> pd.DataFrame:
    """Bin Fare into 4 categories"""
    df = df.copy()
    df['FareBin'] = pd.qcut(df['Fare'], q=4, labels=['Low','Mid','High','VHigh'])
    return df 

def add_age_bins(df: pd.DataFrame) -> pd.DataFrame:
    """Bin Age into child/adult/senior"""
    df['AgeBin'] = pd.cut(df['Age'],bins=[0,12,60,100],labels=['Child','Adult','Senior'])
    return df

def validate_clean(df: pd.DataFrame) -> pd.DataFrame:
    """Assert no nulls remain - fail loudly if something went wrong."""
    null_counts = df.isnull().sum()
    assert null_counts.sum() == 0,\
        f"Null Remain: {null_counts[null_counts>0]}"
    print(f"Clean DataFrame: {df.shape}, 0 nulls")
    return df

# Combine everything in one chain

df_final = (
    pd.read_csv('titanic.csv')
    .drop(columns=['Cabin','Ticket','PassengerId'])
    .dropna(subset=['Embarked'])
    .assign(
        Age = lambda df: df['Age'].fillna(df['Age'].median()),
        Title=lambda df: df['Name'].str.extract(r',\s([^\.]+)\.')[0],
        Sex=lambda df: df['Sex'].map({'male': 0, 'female': 1}),
        FamilySize=lambda df: df['SibSp'] + df['Parch'] + 1,
        IsAlone=lambda df: (df['SibSp'] + df['Parch'] == 0).astype(int),
    )
    .drop(columns=['Name'])
    .pipe(add_fare_bins)
    .pipe(add_age_bins)
    .pipe(validate_clean)
    .reset_index(drop=True)
)

print(df_final.dtypes)
print(df_final.head())
```
