### Vectorization vs Loops

```python
import numpy as np
import time

# PROBLEM 1: Square each element in an array of 1 million numbers

data = np.random.randn(1_000_000)

# Loop version
start = time.time()
result_loop = []
for x in data:
    result_loop.append(x ** 2)
loop_time = time.time() - start
print(f"Loop: {loop_time:.4f} seconds")

# Vectorized version 
start = time.time()
result_vec = data ** 2
vec_time = time.time() - start
print(f"Vectorized: {vec_time:.4f} seconds")

print(f"Speedup: {loop_time / vec_time:.1f}x faster")

# Verify same result
print("Same result:", np.allclose(result_loop, result_vec))
```

```python
# PROBLEM 2: Normalize an array (subtract mean, divide by std)

data = np.random.randn(500_000) # just leave the cell running and let it finish 
                                # you will be shocked to see the results

# loops
result = [] 
start = time.time()
for x in data:
    result.append((x-np.mean(data))/np.std(data))
loop_time = time.time()-start
print(f"Loop: {loop_time:.4f} seconds")

# vectorized 
start = time.time()
result = (data-np.mean(data))/np.std(data)
vec_time = time.time()-start
print(f"Vectorized: {vec_time:.4f} seconds")


print(f"Speedup: {loop_time / vec_time:.1f}x faster")
```

```python
# PROBLEM 3: Count how many elements are above the mean

data = np.random.randn(500_0000)

mean = np.mean(data)

# loops
start = time.time()
count = 0
for i in data:
    if i>mean:
        count += 1
print('Loops - total elements above mean :{}'.format(count))
loop_time = time.time()-start
print(f"Loop: {loop_time:.4f} seconds")

# vectorized
start = time.time()
count = data[data>mean].size
print('Vectorized - total elements above mean :{}'.format(count))
vec_time = time.time()-start
print(f"Vectorized: {vec_time:.4f} seconds")


print(f"Speedup: {loop_time / vec_time:.1f}x faster")

```

```python
# PROBLEM 4: Compute dot product of two arrays manually

arr1 = np.random.randn(5_000_000)
arr2 = np.random.randn(5_000_000)

# loops
start = time.time()
result = 0
for i,j in zip(arr1,arr2):
    result += i*j
print('dot product using loops: {:.2f}'.format(result))

loop_time = time.time()-start
print(f"Loop: {loop_time:.4f} seconds")

# vectorized
start = time.time()
result = np.sum(arr1*arr2)
print('dot product using vectorized: {:.2f}'.format(result))

vec_time = time.time()-start
print(f"Vectorized: {vec_time:.4f} seconds")


print(f"Speedup: {loop_time / vec_time:.1f}x faster")
```

```python
# PROBLEM 5: Find the row with the maximum sum in a 2D matrix

data_2d = np.random.randn(100000, 50)

# loops
start = time.time()

ind = 0
max_sum = 0
for i in range(len(data_2d)):
    rw_sm = sum(data_2d[i])
    if max_sum < rw_sm:
        max_sum = rw_sm
        ind = i

print('Loops - row number {} have max sum {:.2f}'.format(ind,max_sum))

loop_time = time.time()-start
print(f"Loop: {loop_time:.4f} seconds")

# vectorized
start = time.time()
sums = np.sum(data_2d,axis=1)
ind = np.argmax(sums)
max_sum = sums[ind]

print('Vectorized - row number {} have max sum {:.2f}'.format(ind,max_sum))

vec_time = time.time()-start
print(f"Vectorized: {vec_time:.4f} seconds")


print(f"Speedup: {loop_time / vec_time:.1f}x faster")
```

### Broadcasting 

```python
# Broadcasting Rule: NumPy expands smaller arrays to match larger ones
# The rule: dimensions are compared RIGHT to LEFT
# Compatible if: they are equal, or one of them is 1

# Example 1: scalar + array (simplest case)
arr = np.array([1, 2, 3, 4])
result = arr + 10     # 10 is broadcast to [10, 10, 10, 10]
print(result)         # [11, 12, 13, 14]

# Example 2: column vector + row vector (most confusing case)
col = np.array([[1], [2], [3]])    # shape (3, 1)
row = np.array([10, 20, 30, 40])  # shape (4,) → treated as (1, 4)
result = col + row
print(result.shape)   # PREDICT: what shape?
print(result)         # PREDICT: what values?

# Example 3: Normalize each column of a matrix (real ML use case)
data = np.random.randn(100, 5)    # 100 samples, 5 features
mean = data.mean(axis=0)          # shape (5,) — mean of each column
std = data.std(axis=0)            # shape (5,) — std of each column
normalized = (data - mean) / std  # broadcasting makes this work
print(normalized.shape)           # (100, 5) — same shape
print(normalized.mean(axis=0))    # should be ~[0, 0, 0, 0, 0]
print(normalized.std(axis=0))     # should be ~[1, 1, 1, 1, 1]

# INTENTIONALLY break broadcasting
try:
    arr1 = np.ones((3, 4))
    arr2 = np.ones((3, 3))
    result = arr1 + arr2   # This will fail — why?
except ValueError as e:
    print(f"Error: {e}")
```

### Vectorization 10-problems

```python
# PROBLEM 1: Euclidean distance between two points

p1 = np.random.randint(1,10000,1_000_000).astype(np.float64)
p2 = np.random.randint(1,10000,1_000_000).astype(np.float64)

# loops
start = time.time()
distance = sum([(x-y)**2 for x,y in zip(p1,p2)])**0.5   
loop_time = time.time()-start
print('distance between points using loops: {:.2f}'.format(distance))
print(f"Loop: {loop_time:.4f} seconds")

# vectorized 
start = time.time()
distance = np.linalg.norm(p1-p2)
vec_time = time.time()-start
print('distance between points using vectorization: {:.2f}'.format(distance))
print(f"Vectorized: {vec_time:.4f} seconds")


print(f"Speedup: {loop_time / vec_time:.1f}x faster")

```

```python
# PROBLEM 2: Min-max normalization of array
# Formula: (x - min) / (max - min)

array = np.random.randn(5_0000_000)

# loops
start = time.time()

mx = max(array)
mn = min(array)
results = []
for num in array:
    results.append((num-mn)/(mx-mn))

loop_time = time.time()-start
print('array normalized using loops.')
print(f"Loop: {loop_time:.4f} seconds")

# vectorized 
start = time.time()
results = (array - np.min(array))/(np.max(array) - np.min(array))
vec_time = time.time()-start
print('array normalized using vectorization.')
print(f"Vectorized: {vec_time:.4f} seconds")


print(f"Speedup: {loop_time / vec_time:.1f}x faster")
```

```python
data = np.random.randn(1_000_000)

# loop
data_loop = data.copy()
start = time.time()

for ind, val in enumerate(data_loop):
    if val < 0:
        data_loop[ind] = 0

loop_time = time.time() - start

# vectorized
data_vec = data.copy()
start = time.time()

data_vec[data_vec < 0] = 0

vec_time = time.time() - start

print(f"Loop: {loop_time:.4f} seconds")
print(f"Vectorized: {vec_time:.4f} seconds")
print(f"Speedup: {loop_time / vec_time:.1f}x faster")
```

```python
# PROBLEM 7: Compute correlation between two arrays manually

x = np.random.randn(10_000_000)
y = 2 * x + np.random.randn(10_000_000) * 0.5

# loop
start = time.time()

N = len(x)
x_mean = sum(x)/N 
y_mean = sum(y)/N
cov_loop = sum((x-x_mean)*(y-y_mean))/(N-1)
stdx_loop = (sum((x-x_mean)**2)/(N-1))**0.5
stdy_loop = (sum((y-y_mean)**2)/(N-1))**0.5
corrcoef_loop = cov_loop / (stdx_loop*stdy_loop)

loop_time = time.time() - start


# vectorized
start = time.time()
corrcoef_vect = np.corrcoef(x,y)
vect_time = time.time()-start


print('correcoef :- \nLoop - {}\nVectorization - {}'.format(corrcoef_loop,corrcoef_vect[0][1]))

print(f"Loop: {loop_time:.4f} seconds")
print(f"Vectorized: {vec_time:.4f} seconds")
print(f"Speedup: {loop_time / vec_time:.1f}x faster")


```

```python
# PROBLEM 8: One-hot encode a categorical array

labels = np.random.randint(0, 10, size=10_000_000)  # 10 classes

# loop
start = time.time()

num_classes = 10

onehot_loop = np.zeros((len(labels), num_classes), dtype=int)

for i in range(len(labels)):
    onehot_loop[i][labels[i]] = 1

loop_time = time.time() - start


# vectorized
start = time.time()

onehot_vect = np.eye(num_classes, dtype=int)[labels]

vect_time = time.time() - start


print('One-hot encoding completed.')

print(f"Loop: {loop_time:.4f} seconds")
print(f"Vectorized: {vect_time:.4f} seconds")
print(f"Speedup: {loop_time / vect_time:.1f}x faster")
```

```python
# PROBLEM 9: Moving average with window size 3

data = np.random.randn(10_000_000)

# loop
start = time.time()

window = 3

movingavg_loop = []

for i in range(len(data) - window + 1):
    avg = sum(data[i:i+window]) / window
    movingavg_loop.append(avg)

movingavg_loop = np.array(movingavg_loop)

loop_time = time.time() - start


# vectorized
start = time.time()

movingavg_vect = np.convolve(
    data,
    np.ones(window) / window,
    mode='valid'
)

vect_time = time.time() - start


print(
    '\nMoving Average :- \n'
    'Loop - \n{}\n'
    'Vectorization - \n{}'.format(
        movingavg_loop[:10],
        movingavg_vect[:10]
    )
)

print(f"Loop: {loop_time:.6f} seconds")
print(f"Vectorized: {vect_time:.6f} seconds")
print(f"Speedup: {loop_time / vect_time:.1f}x faster")
```

```python
# PROBLEM 10: Clip array values between -2 and 2
data = np.random.randn(1_000_000) * 5

# loop
start = time.time()

clipped_loop = np.empty_like(data)

for i in range(len(data)):

    if data[i] < -2:
        clipped_loop[i] = -2

    elif data[i] > 2:
        clipped_loop[i] = 2

    else:
        clipped_loop[i] = data[i]

loop_time = time.time() - start


# vectorized
start = time.time()

clipped_vect = np.clip(data, -2, 2)

vect_time = time.time() - start


print(
    '\nClipped Array :- \n'
    'Loop - \n{}\n'
    'Vectorization - \n{}'.format(
        clipped_loop[:10],
        clipped_vect[:10]
    )
)

print(f"Loop: {loop_time:.6f} seconds")
print(f"Vectorized: {vect_time:.6f} seconds")
print(f"Speedup: {loop_time / vect_time:.1f}x faster")
```

### Broadcasting Deep Dive + Edge Cases

```python
# REAL ML USE CASE: Subtract mean from each row (not column)
data = np.random.randn(5, 3)   # 5 samples, 3 features
row_means = data.mean(axis=1)  # shape (5,) — mean of each row
print(row_means.shape)

# This will FAIL — why?
# result = data - row_means

# Fix: reshape row_means to (5, 1) so broadcasting works
result = data - row_means.reshape(-1, 1)   # or row_means[:, np.newaxis]
print(result.shape)   # (5, 3) — correct

# REAL ML USE CASE: Cosine similarity between two matrices
# (Used in recommendation systems and NLP)
a = np.random.randn(10, 50)   # 10 query vectors, 50 dimensions
b = np.random.randn(20, 50)   # 20 document vectors, 50 dimensions

# Normalize each row to unit length
a_norm = a / np.linalg.norm(a, axis=1, keepdims=True)
b_norm = b / np.linalg.norm(b, axis=1, keepdims=True)

# Compute all pairwise similarities
similarity = a_norm @ b_norm.T    # shape (10, 20)
print(similarity.shape)           # 10 queries × 20 documents
print("Max similarity:", similarity.max())
print("Min similarity:", similarity.min())
```
